```python
import pandas
df = pandas.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv")
df.to_sql("CHICAGO_PUBLIC_SCHOOLS_DATA", con, if_exists='replace', index=False, method="multi")
print(df.head())
```

```python
%sql select NAME_OF_SCHOOL,SAFETY_SCORE from CHICAGO_PUBLIC_SCHOOLS_DATA where SAFETY_SCORE= (select max(SAFETY_SCORE)as highest_score from CHICAGO_PUBLIC_SCHOOLS_DATA )
```

```python
%sql select NAME_OF_SCHOOL,AVERAGE_STUDENT_ATTENDANCE from CHICAGO_PUBLIC_SCHOOLS_DATA \
order by  AVERAGE_STUDENT_ATTENDANCE\
limit 5
```

```python
%sql select NAME_OF_SCHOOL,AVERAGE_STUDENT_ATTENDANCE from CHICAGO_PUBLIC_SCHOOLS_DATA \
order by  AVERAGE_STUDENT_ATTENDANCE desc \
limit 10
```

```python
%sql select NAME_OF_SCHOOL,replace("AVERAGE_STUDENT_ATTENDANCE","%", '')as highest_score from CHICAGO_PUBLIC_SCHOOLS_DATA \
order by  AVERAGE_STUDENT_ATTENDANCE desc \
limit 10
```

```python
%sql SELECT Name_of_school,AVERAGE_STUDENT_ATTENDANCE from CHICAGO_PUBLIC_SCHOOLS_DATA \
WHERE cast (replace(AVERAGE_STUDENT_ATTENDANCE,'%','')as double)<70\
order by AVERAGE_STUDENT_ATTENDANCE
```

```python
#%sql select name from PRAGMA_TABLE_INFO("CHICAGO_PUBLIC_SCHOOLS_DATA")
%sql select COMMUNITY_AREA_NAME,sum(COLLEGE_ENROLLMENT) from CHICAGO_PUBLIC_SCHOOLS_DATA\
group by COMMUNITY_AREA_NAME
```

```python
%sql SELECT COMMUNITY_AREA_NAME,sum(COLLEGE_ENROLLMENT)as total_enrollment from CHICAGO_PUBLIC_SCHOOLS_DATA\
group by COMMUNITY_AREA_NAME \
ORDER BY total_enrollment asc
```

```python
%sql select NAME_OF_SCHOOL,SAFETY_SCORE FROM CHICAGO_PUBLIC_SCHOOLS_DATA \
WHERE SAFETY_SCORE != "none" \
ORDER BY SAFETY_SCORE  \
limit 5
```
